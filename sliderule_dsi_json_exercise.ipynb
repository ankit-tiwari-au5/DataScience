## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
import pandas as pd
import json
from pandas.io.json import json_normalize
```

```python
# Read json file and load as Pandas dataframe
wb_df = pd.read_json('data/world_bank_projects.json')
wb_df.head()
```

```python
#1. Find the 10 countries with most projects
countries = wb_df['countryname'].value_counts()

#Count number of null values in 'countries'
countries.isnull().sum()


#get top 10 countries
countries[:10]
```

```python
#2. Find the top 10 major project themes (using column 'mjtheme_namecode')

#Load json file
data = json.load((open('data/world_bank_projects.json')))

#Use json_normalize to flatten nested json
themes = json_normalize(data, 'mjtheme_namecode')

#type(themes)
top_theme = themes.loc[themes['code'].value_counts()]

#get top 10
top_theme[:10]
#code 11 has missing values
```

```python
top_theme.isnull().sum()
```

```python
#code 11 has missing values
#check missing values
miss = themes[themes.code == '11']
#get value of missing theme name
txt = miss.name.unique()[1]
#check type of value in sample of missing 'name' column
type(txt)
#Print missing value name
print(txt)
```

```python
#Fill in missing values for code '11' with 'Environment and natural resources management' using .replace()
#Create a top 10 major project themes dataframe with the missing names filled in.
top_theme.replace({'name': ''}, {'name': txt}, inplace=True)
top_theme
```
